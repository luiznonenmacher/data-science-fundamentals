# Visualizing Earning Based on College Majors

### Introduction

This project will analyse the job outcomes of students who graduated from college between 2010 and 2012. The original data was released by American Community Survey, but we will use a cleaned dataset found on [FiveThirtyEight Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

The learning objective of this project is plotting in Jupyter Notebook using Pandas and matplotlib. 

Each row in dataset represents a different major. Here are the most important columns in the dataset:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

### Histograms

We will construct 8 histograms to better understand the distribution of the following columns:

- Sample_size
- Median
- Employed
- Full-time
- ShareWomen
- Unemployment_rate
- Men
- Women

```python
recent_grads['Sample_size'].hist(bins=20)
```

The majority of the courses have a small sample size, with the majority having less than 200 samples.

```python
recent_grads['Median'].hist(bins=20)
```

Most of the courses median salaries are on the left size of the curve, with a peak around $ 30,000. There are just a few courses with median salaries higher than $ 50,000.

```python
recent_grads['Employed'].hist(bins=30)
```

```python
recent_grads['Full_time'].hist(bins=25)
```

Looking at those histograms, it appears that most courses don't have lots of persons employed and full-time jobs, but the problem with this analysis is that the numbers are absolute, not relative, so the data is inconclusive.

```python
recent_grads['ShareWomen'].hist(bins=30)
```

The share of women is wel distributed between courses, with some courses predominantly male and others predominantly female.

The most frequent bin is around 0,42, representing more men than women. On the other side, if we count and compare the number of courses where more than 80% is women or men, the quantities are mostrly the same (18 for women, 17 for men). 

```python
recent_grads['Unemployment_rate'].hist(bins=30)
```

The unemployment rate is concentrated around 0.07, but there is another small cluster around 0.10. 

```python
recent_grads['Men'].hist(bins=30)
```

```python
recent_grads['Women'].hist(bins=30)
```

The Men and Women distributions have the same problem of Full Time and Employed: considering absolute numbers, which don't provide much in terms of insights.
